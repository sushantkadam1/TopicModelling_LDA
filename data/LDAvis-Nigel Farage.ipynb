# Topic Modeling - Nigel Farage - Twitter Followers

We use LDAvis to visualize several LDA modeling of the followers of the @BorisJohnson's twitteraccount.

We take 100 raw tweets of 5000 random followers of the politicians for the analysis.

The different LDAs were trained with the following parameters

* 50 topics, 10 passes, alpha = 0.001
* 25 topics, 10 passes, alpha = 0.01
* 10 topics, 10 passes, alpha = 0.001

Extraction of the data from twitter was done via
And the dictionary and corpus were created via 

To see the best results, set lambda between [0.4, 0.5]. Lowering Lambda gives more importance to words that are discriminatory for the active topic, words that best define the topic. 

```python
# Load the corpus and dictionary
from gensim import corpora, models
import pyLDAvis.gensim

corpus = corpora.MmCorpus('Nigel_Farage_tok.mm')
dictionary = corpora.Dictionary.load('Nigel_Farage_tok.dict')
```

# LDA Model with 50 Topics

```python
# First LDA model with 50 topics, 10 passes, alpha = 0.001
lda = models.LdaModel.load('Nigel_Farage_tok50.lda')
followers_data =  pyLDAvis.gensim.prepare(lda, corpus, dictionary, mds='mmds')
pyLDAvis.display(followers_data)
```

1 is about the political landscape in the Unite States of America

3 is again in the political landscape in UK

7 speaks about the labour and conservative party in UK

8 is about the football teams in UK 

10 is about the ongoing protest in Hongkong and its association with China

12 is about the referendum and the Brexit

# LDA Model with 25 Topics

```python
# First LDA model with 25 topics, 10 passes, alpha = 0.001
lda = models.LdaModel.load('Nigel_Farage_tok25.lda')
followers_data =  pyLDAvis.gensim.prepare(lda, corpus, dictionary, mds='mmds')
pyLDAvis.display(followers_data)
```

The topic spread is much better.
And the topics are mostly pretty coherent.

* 1 is about the social web
* 2 is more colloquial about daily life with words like happy, lol, guys, sure, ever, day, ...
* 12 is all about data science with words like dataviz, spark, drivendata, ipython, machinelearning, ...
* 5 is about bitcoins and cryptocurrencies
* 17 most probably comes from French followers that did go through our filtering of non english accounts.
* 29 is about jobs in the UK

The final model has more topics (100) and was allowed to converge more with 100 passes.
