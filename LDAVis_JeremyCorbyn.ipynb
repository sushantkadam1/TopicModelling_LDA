# Topic Modeling of Twitter Followers of Jeremy Corbyn

We use LDAvis to visualize several LDA modeling of the followers of the @JeremyCorbyn twitteraccount.

We take 100 raw tweets of 5000 random followers of the politicians for the analysis.

The different LDAs were trained with the following parameters

* 20 topics, 10 passes, alpha = 0.001
* 10 topics, 10 passes, alpha = 0.001
* 50 topics, 10 passes, alpha = 0.001

Extraction of the data from twitter was done via [this python 2 script](https://github.com/sushantkadam1/TopicModelling_LDA)
And the dictionary and corpus were created via [this one](https://github.com/sushantkadam1/TopicModelling_LDA)

To see the best results, set lambda around [0.5, 0.6]. Lowering Lambda gives more importance to words that are discriminatory for the active topic, words that best define the topic. 

You can skip the 2 first models and jump to the last model which is the best (50 topics)

A working version of this notebook is available on [nbviewer](https://nbviewer.jupyter.org/github/sushantkadam1/TopicModelling_LDA/blob/master/LDAVis_JeremyCorbyn.ipynb#topic=0&lambda=0.51&term=)


```python
# Load the corpus and dictionary
from gensim import corpora, models
import pyLDAvis.gensim

corpus = corpora.MmCorpus('data/jeremycorbyn_tok.mm')
dictionary = corpora.Dictionary.load('data/jeremycorbyn_tok.dict')
```

```python
# First LDA model with 20 topics, 10 passes, alpha = 0.001
lda = models.LdaModel.load('data/jeremycorbyn_tok20.lda')
followers_data =  pyLDAvis.gensim.prepare(lda, corpus, dictionary, mds='mmds')
pyLDAvis.display(followers_data)
```

The topic spread is much better than the 10 that follows below.
And the topics are mostly pretty coherent.

* 1 is about the youth related topics
* 2 is about politics which includes both US and UK,
* 3 is about football(Soccer)
* 4 is about emotions 
* 5 is specifically about US politics
* 6 and 8 are two clusters with two different clusters but in same language = SPANISH language
* 10, 14 and 12 are about Middle-east and Pakistan = URDU language. 14 is closer to Pakistan
* 13 is about INDIAN politics = Hindi + English
* 7 is about gaming and electronics
* 11 and 9 most probably comes from French followers that did go through our filtering of non english accounts.
* 15 is about softwares


```python
# First LDA model with 10 topics, 10 passes, alpha = 0.001
lda = models.LdaModel.load('data/jeremycorbyn_tok10.lda')
followers_data =  pyLDAvis.gensim.prepare(lda, corpus, dictionary, mds='mmds')
pyLDAvis.display(followers_data)
```

With K=10 topics, nearly all the topics are aggregated together and difficult to distinguish. 

```python
# First LDA model with 50 topics, 10 passes, alpha = 0.001
lda = models.LdaModel.load('data/jeremycorbyn_tok50.lda')
followers_data =  pyLDAvis.gensim.prepare(lda, corpus, dictionary, mds='mmds')
pyLDAvis.display(followers_data)
```
